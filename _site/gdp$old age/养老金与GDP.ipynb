```python
df = df_养老金
df.columns = [ int(x.replace('年','')) if x!='地区' else x for x in df.columns]
years = [ x for x in df.columns if x!='地区'] 
df_melted = pd.melt(df, id_vars=['地区'], value_vars=years)
df_melted.columns = ['地区', '年', '养老金']
df_养老金_done = df_melted
df_养老金_done
```

```python
df_p = df_养老金_done.join(df_GDP_done,lsuffix='', rsuffix='_GDP')
df_p
```

```python
df_p.drop(["地区_GDP","年_GDP"], axis=1).set_index(["地区","年"])
```

```python
df_pp = df_p.drop(["地区_GDP","年_GDP"], axis=1).set_index(["地区","年"])
```

```python
df_pp.query("年==2015")
```

```python
%matplotlib inline
df_pp.query("年==2015").plot(kind='scatter', x='养老金', y='GDP', 
          title = '养老金与GDP的关系：分省数据')
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pylab
mpl.rcParams['font.sans-serif']=['SimHei']
mpl.rcParams['axes.unicode_minus']=False

df_plot = df_pp.query("年==2015")
df_plot.plot(kind='scatter', x='养老金', y='GDP', 
          title = '养老金与GDP的关系：分省数据')
```

```python
import numpy as np
fit = np.polyfit(x=df_pp.query("年==2015")['养老金'], y=df_pp.query("年==2015")['GDP'], deg=1)
```

```python
x=df_plot['养老金']
y=df_plot['GDP']
ax = df_plot.plot(kind='scatter', x='养老金', y='GDP', 
          title = '养老金与GDP的关系：分省数据')
ax.plot(x, fit[0] * x + fit[1], color='red')
ax.scatter(x, y)
```

```python
df_pp['desc'] = df_pp.index.get_level_values(0)
```

```python
from bokeh.plotting import figure, output_file, show, output_notebook
from bokeh.models import HoverTool, BoxSelectTool

title = '分省数据' 
xlabel = '养老金：养老金结余' 
ylabel = 'GDP：产值'

hover = HoverTool(
    tooltips=[
        ( '养老金结余',  '$x{0,F}'), 
        ( '产值',  '$y{0,F}'),
        ( '地区',  '@desc'  ),
    ]
)

TOOLS = [hover, "pan, wheel_zoom, box_zoom, reset, resize"]
fig = figure(plot_width=800, plot_height=600,
           x_axis_label = xlabel, y_axis_label = ylabel, 
           title=title, tools=TOOLS)

df_plot = df_pp.query("年==2015")

fig.circle("养老金", "GDP",
         source = df_plot,
         fill_alpha=0.6, line_color=None, size= 30)


xs = [min(df_plot["养老金"]), max(df_plot["养老金"])]
fig.line(x=xs, y=[ fit[0] * x + fit[1] for x in xs] )

output_notebook()
show(fig)
```
